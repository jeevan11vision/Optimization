```python
def f(x):
    fx = 100 * (x[1]  - x[0] ** 2) ** 2 + (1 - x[0] ** 2)
    return fx

def f_(x, y):
    fx = 100 * (y  - x ** 2) ** 2 + (1 - x ** 2)
    return fx
```
